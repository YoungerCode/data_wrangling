#### 2. Create new columns containing the total score of each of the personality test subscales. To do this, write a function (or functions) that will calculate the total score for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion), as set out in scoring. In other words, for the conscientiousness total score, all items marked as belonging to ‘conscientiousness’ should be summed.

```python
def convertData(row, Nums_Cat):
    count = 0
    touch = 0
    for values in personality_score_2.columns:
        lst = row[values]
        if type(lst) == str:
            cat = lst[1]
            result = int(lst[4])
            if cat == Nums_Cat:
                count += result
    return count                   
```

```python
personality_score_2['Conscienctiousness'] = personality_score_2.apply(lambda row : convertData(row,'1'), axis = 1)
personality_score_2['Emotional_Stablility'] = personality_score_2.apply(lambda row : convertData(row,'2'), axis = 1)
personality_score_2['Openness'] = personality_score_2.apply(lambda row : convertData(row,'3'), axis = 1)
personality_score_2['Agreeable'] = personality_score_2.apply(lambda row : convertData(row,'4'), axis = 1)
personality_score_2['Extraversion'] = personality_score_2.apply(lambda row : convertData(row,'5'), axis = 1)
```
