###  1. Import the dataset personality_scores.csv. Examine the data frame for duplicates (based on ID), and drop any duplicates that exist. Use an assert statement to check that the new data frame is the length of the unique entries of the original data frame.

#### Importing  datasets and libraries

```python
import pandas as pd
from pandas.util.testing import assert_frame_equal
import numpy as np
```

```python
personality_score = pd.read_csv("personality_scores.csv", sep = ";")
departments = pd.read_csv("departments.csv", sep = ";")
```

#### Making a copy of a dataframe to dropping duplicates and null values, Also used  assert_frame_equal to compare the two DataFrames



```python
personality_score_2 = personality_score.copy()
```

```python
personality_score_2.drop_duplicates(subset ="ID", 
                     keep = False, inplace = False)
```

```python
personality_score_2.dropna(axis=1, how='all', inplace=True)
```

```python
assert_frame_equal(personality_score, personality_score_2)
```

### 2. Create new columns containing the total score of each of the personality test subscales. To do this, write a function (or functions) that will calculate the total score for each of the subscales (conscientiousness, emotional stability, openness to new experience, agreeableness, extraversion), as set out in scoring. In other words, for the conscientiousness total score, all items marked as belonging to ‘conscientiousness’ should be summed.

#### Function that tallies the scores of subscales and used lambda function to assign sums to the created columns.

```python
def convertData(row, Nums_Cat):
    count = 0
    touch = 0
    for values in personality_score_2.columns:
        lst = row[values]
        if type(lst) == str:
            cat = lst[1]
            result = int(lst[4])
            if cat == Nums_Cat:
                count += result
    return count                   
```

```python
personality_score_2['Conscienctiousness'] = personality_score_2.apply(lambda row : convertData(row,'3'), axis = 1)
personality_score_2['Emotional_Stablility'] = personality_score_2.apply(lambda row : convertData(row,'4'), axis = 1)
personality_score_2['Openness'] = personality_score_2.apply(lambda row : convertData(row,'5'), axis = 1)
personality_score_2['Agreeable'] = personality_score_2.apply(lambda row : convertData(row,'2'), axis = 1)
personality_score_2['Extraversion'] = personality_score_2.apply(lambda row : convertData(row,'1'), axis = 1)
```

### 4.Filter the merged data frame so that you get only the applicants who scored less than 30 on emotional stability, conscientiousness AND agreeableness. Print the ID numbers and departments of these applicants to the screen, and also assign these applicants the tag “high_risk” in a new column. All other applicants get the tag “low_risk”

#### Creating a new dataframe called Risk_status and it measures risk status based on the given conditions

```python
Risk_status =pd.DataFrame(personality_score_2_merged[['ID','Conscienctiousness','Emotional_Stablility','Openness','Agreeable','Extraversion','Department']])
```

```python
Risk_status
```

```python
Risk_status['risk_status'] = np.where(
    (Risk_status['Emotional_Stablility'] < 30) & \
    (Risk_status.Conscienctiousness < 30) & (Risk_status.Agreeable < 30), 'High risk', 'Low risk'
)
Risk_status.head()
```

### 4. Create a new data frame with a count of the number of low and high risk applicants within each department. Let each department be a separate column. 



```python
# we are calling the columns, grouping them and making them into a DataFrame 
summary = Risk_status[['risk_status', 'Department']].groupby(['risk_status', 'Department']).size()
pd.DataFrame(summary)
```
